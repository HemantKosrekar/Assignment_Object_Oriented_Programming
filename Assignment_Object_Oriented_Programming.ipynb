````python
# 7. Write a code that generates palindromic numbers up to n using a generator.

def gen_palindromic_num(num):
    i = 1
    count = 1
    while True:
        if count > num:
            break
        if str(i) == str(i)[::-1]:
            yield i
            count += 1
        i += 1

generator_obj = gen_palindromic_num(15)

for i,palindromic_num in enumerate(generator_obj,start=1):
    print(f"{i} :",palindromic_num)
````

````python
# 8. Write a code that generates even numbers from 2 to n using a generator.

def gen_even_num(num):
    count = 1
    i = 1
    while True:
        if i > num:
            break

        if count % 2 == 0:
            yield count
            i += 1
        count += 1
        
even_num_generator = gen_even_num(10)

for i, even_num in enumerate(even_num_generator, start=1):
    print(f"{i} :", even_num)
            
````

````python
# 10. Write a code that generates prime numbers up to n using a generator.

def is_prime(n):
    if n == 1:
        return False
    for d in range(2, n):
        if n % d == 0:
            return False
    return True

def gen_prime_num(num):
    count = 1
    i = 1
    while True:
        if i > num:
            break

        if is_prime(count):
            yield count
            i += 1
        count += 1

prime_num_generator = gen_prime_num(10)

for i, prime_num in enumerate(prime_num_generator, start=1):
    print(f"{i} :", prime_num)
````

#### 29. What is inheritance in OOP?
- It's phenomenon in object oriented programming where new class(sub class) acquires properties and methods of existing class(parent class).  

- There are 5 types of inheritance in python  
    1. single level inheritance  
    2. multi level inheritance  
    3. multiple inheritance  
    4. hierarchical inheritance  
    5. hybrid inheritance   

1. single level inheritance  
    - A derived class inherits from one base class.  
    - Ex:-  
    	A --> B(A)  
    	Here A is parent class and B is child class, B derived from A.  

2. multi level inheritance  
    - It's an inheritance chain where a class derives from a class that is itself derived from another class.  
    - Ex:-  
    	A --> B(A) --> C(B)  
    	Here class B derived from class A and class C derived from class B.  
	

3. multiple inheritance  
    - A derived class inherits from more than one parent class.  
    - Ex:-  
    	A --> C(A,B) <-- B  
    	here class C is derived from multiple parent class A and class B  

4. hierarchical inheritance  
    - In this inheritance several child classes derive from a single parent class  
    - Ex:-  
    	 B(A) <-- A --> C(A)  
    	 here A is parent class and class B & class C both are derived from parent class A  

5. hybrid inheritance  
    - It's an combination of one or more inheritance types.  
    - here class B and class C are derived from parent class A that is hierarchical inheritance  
    and further class D is derived from multiple parent class B & class C that is multiple inheritance  
    - It's combination of multiple inheritance types that's why it's called hybrid inheritance.
    
       Ex:-
    
       ```
            A
           / \
        B(A)  C(A)
           \ /
          D(B,C)
       ```

#### 39. How are abstract methods different from regular methods in Python?

| Aspect                  | Regular Method                                | Abstract Method                                    |
|-------------------------|----------------------------------------------|---------------------------------------------------|
| Definition              | Method with a complete implementation (body). | Method declared without implementation (no body). |
| Purpose                 | Performs specific tasks directly.             | Serves as a template to be implemented by subclasses. |
| Location                | Can be in any class (regular or abstract).    | Must be declared in an abstract class or interface. |
| Instantiation           | Can be called on instances directly.          | Cannot be called until implemented in subclass.   |
| Implementation in subclass | Optional to override; inherits default behavior if not overridden. | Mandatory to implement in any concrete subclass.   |
| Object creation         | Class containing regular methods can be instantiated. | Class with abstract methods cannot be instantiated directly. |
| Usage                   | Provides concrete functionality.               | Enforces a contract for subclasses to provide functionality. |


#### 40. How can you achieve abstraction using interfaces in Python?
- You achieve abstraction using interfaces by defining an abstract base class with abstract methods using the "abc" module.  
- This abstract class acts as a blueprint with method names but no implementation  
- Any class inheriting from it must provide its own implementation for these abstract methods.  
- This ensures that all subclasses follow a consistent structure while hiding the details of how the methods work internally.  
- In simple terms, interfaces in Python specify what methods a class should have, leaving the actual work to the subclasses.
- This promotes code consistency and enforces that all subclasses follow the same interface while hiding internal details, which is the essence of abstraction.

````python
# 41. Can you provide an example of how abstraction can be utilized to create a common interface for a group of related classes in Python?

'''
In below example [Shape] is an abstract class and [printarea] is an abstract method.
'''

from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def printarea(self):
        pass
class Rectangle(Shape):
    type = "Rectangle"
    sides = 4

    def __init__(self, width, height):
        self.width = width
        self.height = height

    def printarea(self):
        area = self.width * self.height
        print(f"Rectangle area: {area}")

obj1 = Rectangle(10, 20)
obj1.printarea()

````

#### 42. How does Python achieve polymorphism through method overriding?
- Python achieves polymorphism through method overriding by allowing a subclass to provide a specific implementation of a method that is already defined in its parent class.  
-  When the same method is called on an object, Python determines at runtime which version of the method to execute based on the actual type of the object, enabling different behaviors for the same method name.  
- For example, a base class Animal may have a method called sound(), while its subclasses Dog and Cat override this method to return "Bark" and "Meow" respectively; calling the sound() method on any Animal reference will produce the correct output based on the object's true class.  
- This runtime decision-making for method calls enhances flexibility and supports dynamic method dispatch, which is a core aspect of polymorphism in Python.

#### 47. How do you achieve encapsulation in Python?
- Encapsulation in Python is achieved by restricting access to certain parts of an object, usually by making some variables or methods private so they cannot be accessed directly from outside the class.   
- This is done by prefixing variable or method names with double underscores (__). Encapsulation helps protect the internal state of an object and controls how that state is changed, which improves security and prevents accidental interference.  
- To allow controlled access, public getter and setter methods are used that can include validation before modifying or returning the private data. For example, a class can have a private attribute __salary that is accessed and updated only through methods like get_salary() and set_salary().  
- This makes the code more organized, secure, and easier to maintain, which is the essence of encapsulation in Python.  

#### 48. Can encapsulation be bypassed in Python? If so, how?
- Yes, encapsulation in Python can be bypassed because Python’s private variables are only protected by a convention called name mangling, not by strict access control like in some other programming languages.  
- When you prefix a variable name with double underscores (e.g., __data), Python internally changes the name to include the class name, making it harder but not impossible to access directly from outside the class.  
-  A user can still access these "private" variables by using the modified name (e.g., _ClassName__data), bypassing the encapsulation protection.   
- This design allows flexibility for developers but relies on discipline to respect encapsulation rules rather than enforcing them strictly.   
- So, while encapsulation improves data hiding, it can still be bypassed if needed, though that is generally discouraged because it breaks the intended protection.

````python
# 56. Create a decorator that preserves the metadata of the original function.

import functools

def my_decorator(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        print(f"Before executing {func.__name__}")
        print(f"After executing {func.__name__}")
        return func(*args, **kwargs)
    return wrapper

@my_decorator
def say_hello(name):
    """Greet a person by name."""
    print(f"Hello, {name}!")

print("Function Name:",say_hello.__name__) 
print("Doc string:",say_hello.__doc__)

say_hello("Hemant")
````

#### 63. How does the __len__() method work in Python? Provide an example.
- The __len__() method in Python is a special method used to define how the built-in len() function calculates the length of an object.
-  When you call len() on an object, Python internally calls the object's __len__() method to get and return the length as a non-negative integer.
- This method allows you to customize what "length" means for your own classes.

````python
# __len__() method example

class MyList:
    def __init__(self, items):
        self.items = items

    # custom legnth function
    def __len__(self):
        len = 0
        for _ in self.items:
            len += 1
        return len
    

my_list = MyList([1, 2, 3, 4, 5])
print(len(my_list))
````

#### 64. Explain the usage of the __add__() method in Python classes. Provide an example.
- The ```__add__()``` method in Python classes is a special (magic or dunder) method used to define the behavior of the addition operator + for objects of the class.  
- When you use the + operator between two objects, Python internally calls the left object's ```__add__()``` method with the right object as the argument.
- By implementing this method, you can customize how instances of your class are added together, often returning a new instance representing the combined result.

````python
# __add__() method example

class Complex:
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag

    def __str__(self):
        return f"{self.real} + {self.imag}j"

    # customized __add__ method for complex number
    def __add__(self,complex_num):
        self.real = self.real + complex_num.real
        self.real = self.imag + complex_num.imag
        return Complex(self.real,self.imag)

c1 = Complex(2, 3)
c2 = Complex(3, 4)
c3 = c1 + c2
print(type(c3))
print(c3)
        
    
````

#### 65. What is the purpose of the __getitem__() method in Python? Provide an example.

- The ```__getitem__()``` method in Python is a special (magic) method that allows an object to support indexing using the square bracket notation, like lists or dictionaries.
- When you use ```obj[key]```, Python internally calls obj.```__getitem__(key)```.
-  By implementing this method in your class, you can customize how your objects respond to item access, enabling you to define behavior for retrieving items using keys or indices.

````python
# __getitem__() method implementation

class MyList:
    def __init__(self, items):
        self.items = items

    def __getitem__(self, index):
        return self.items[index]

my_list = MyList([10, 20, 30, 40])
print(my_list[3])
````

#### 67. What is the purpose of a getter method in Python? Provide an example demonstrating the use of a getter method using property decorators.

- The purpose of a getter method in Python is to provide controlled access to a private attribute of a class. 
-  Using the @property decorator, a getter method allows you to access the value of an attribute like a regular attribute, but actually calls a method behind the scenes.
- This is useful for encapsulation, allowing you to add logic when retrieving the value without changing how the attribute is accessed externally.

````python
# implementation of property decorator in python

class BankAccount:
    def __init__(self,acc_num):
        self.__acc_num = acc_num

    @property
    def acc_num(self):
        return self.__acc_num

emp1 = BankAccount("123456")
print(emp1.acc_num) # you can access account number as an attribute but can't change it
````

#### 68. Explain the role of setter methods in Python. Demonstrate how to use a setter method to modify a class attribute using property decorators.
- The role of setter methods in Python is to provide controlled ways to modify the value of a private or protected class attribute.
- They are used to add validation, logging, or other processing whenever an attribute is set, helping to enforce data integrity.
- Using the @property decorator together with a setter method ```(@property_name.setter)```, Python lets you define this behavior while allowing attribute access syntax that is clean and intuitive.

````python
# example of setter method in python

class Person:
    def __init__(self, age):
        self._age = age  # private attribute

    @property
    def age(self):
        # Getter method
        return self._age

    @age.setter
    def age(self, value):
        # Setter method
        if value < 0:
            print("Age cannot be negative.")
        else:
            self._age = value

person = Person(25)
print(person.age)

person.age = 30    # Updates age to 30
print(person.age) 
person.age = -5    # Attempt to set invalid age
print(person.age)
````

#### 70. Explain the use of the @deleter decorator in Python property decorators. Provide a code example demonstrating its application.
- The ```@deleter``` decorator in Python is used within the property decorator structure to define a method that handles the deletion of a property. 
- When you use ```@property_name.deleter```, you specify custom behavior that should occur when the property is deleted with the del keyword.
- This helps manage resource cleanup or enforce rules when a property value is removed, adding control and logic to attribute deletion.

````python
# implementation of @deleter decorator in python

class Person:
    def __init__(self, name):
        self._name = name  # private attribute

    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, value):
        self._name = value

    @name.deleter
    def name(self):
        print("Deleting name...")
        del self._name

person = Person("Alice")
print(person.name)  
person.name = "Bob"
print(person.name) 
del person.name
# print(person.name)  #It Would raise AttributeError because the attribute is deleted
````

#### 71. How does encapsulation relate to property decorators in Python? Provide an example showcasing encapsulation using property decorators.
- Encapsulation in Python means restricting direct access to some of an object's components to protect the integrity of the data. 
-  Property decorators ```(@property, @setter, @deleter)``` are closely related to encapsulation because they allow controlled access to private attributes in a class. 
-  Instead of exposing private variables directly, you use property decorators to define getter, setter, and deleter methods that manage how the attribute is accessed or modified. 
- This keeps the attribute hidden while providing a clean and controlled interface for external code.

````python
# example showcasing encapsulation using property decorators

class Person:
    def __init__(self, name, age):
        self.__name = name     # Private attribute
        self.__age = age       # Private attribute

    @property
    def name(self):
        # Getter for name
        return self.__name

    @name.setter
    def name(self, value):
        # Setter for name with validation
        if not value:
            raise ValueError("Name cannot be empty")
        self.__name = value

    @property
    def age(self):
        # Getter for age
        return self.__age

    @age.setter
    def age(self, value):
        # Setter for age with validation
        if value < 0:
            raise ValueError("Age cannot be negative")
        self.__age = value

person = Person("Amit", 32)
print(person.name)  # Accesses name with getter
print("person age before :",person.age)
person.age = 35     # Modifies age with setter
print("person age after :",person.age)

# person.name = ""    # This would raise a ValueError due to validation
# person.age = -5     # This would also raise a ValueError
````
