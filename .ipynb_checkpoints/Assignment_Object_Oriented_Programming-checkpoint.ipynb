```python
# Example of args

def sum_of_num(*args):
    print(type(args)) #it's a tuple
    num = 0
    for i in args:
        num += i
    return num

sum_of_num(5,3,2,1)
```

```python
# Example of kwargs

def show_student_info(**kwargs):
    print(type(kwargs)) # it's a dictionary

    for key,value in kwargs.items():
        print(f"{key} : {value}")

show_student_info(name="Mahesh",course="data science",platform="pwskills")
```

```python
# 6. Write a code that generates the squares of numbers from 1 to n using a generator.

def square_of_numbers(num):

    for i in range(1, num+1):
        yield i**2

gen_square = square_of_numbers(12) # generator

for square_num in gen_square:
    print(square_num)
```

```python
# 7. Write a code that generates palindromic numbers up to n using a generator.

def gen_palindromic_num(num):
    i = 1
    count = 1
    while True:
        if count > num:
            break
        if str(i) == str(i)[::-1]:
            yield i
            count += 1
        i += 1

generator_obj = gen_palindromic_num(15)

for i,palindromic_num in enumerate(generator_obj,start=1):
    print(f"{i} :",palindromic_num)
```

```python
# 8. Write a code that generates even numbers from 2 to n using a generator.

def gen_even_num(num):
    count = 1
    i = 1
    while True:
        if i > num:
            break

        if count % 2 == 0:
            yield count
            i += 1
        count += 1
        
even_num_generator = gen_even_num(10)

for i, even_num in enumerate(even_num_generator, start=1):
    print(f"{i} :", even_num)
            
```

```python
# 9. Write a code that generates powers of two up to n using a generator.

num = int(input("Enter the num : "))

# creating generator using generator expression
gen_power_of_two = (i**2 for i in range(1,num)) 

for i, power_num in enumerate(gen_power_of_two, start=1):
    print(f"{i} :", power_num)

    
```

```python
# 10. Write a code that generates prime numbers up to n using a generator.

def is_prime(n):
    if n == 1:
        return False
    for d in range(2, n):
        if n % d == 0:
            return False
    return True

def gen_prime_num(num):
    count = 1
    i = 1
    while True:
        if i > num:
            break

        if is_prime(count):
            yield count
            i += 1
        count += 1

prime_num_generator = gen_prime_num(10)

for i, prime_num in enumerate(prime_num_generator, start=1):
    print(f"{i} :", prime_num)
```

```python
# 18. Write a code that calculates the product of positive numbers from a given list.

def product_of_positive_num(ls):
    output_num = 1
    for i in ls:
        if isinstance(i,int) and i > 0:
            output_num = output_num * i

    return output_num

input_ls = [1,2,3,4,5,-1,-2,-3,-4]
print("product_of_positive_num :",product_of_positive_num(input_ls))
```

```python
# 19. Write a code that doubles the values of odd numbers from a given list.

def double_the_num(ls):
    output_ls = []
    for i in ls:
        if isinstance(i,int) and i % 2 == 0:
            output_ls.append(i * 2)

    return output_ls

input_ls = [1,2,3,4,5,6,7,8,9]
print("output list :",double_the_num(input_ls))
```

```python
# 21. Write a code that filters out prime numbers from a given list.

def is_prime(n):
    if n == 1:
        return False
    for d in range(2, n):
        if n % d == 0:
            return False
    return True

filtered_prime = lambda ls: [i for i in ls if is_prime(i)]
input_ls = [1,2,3,4,5,6,7,8,9]
print("filtered prime list: ",filtered_prime(input_ls))
```
